```python
textlist=os.listdir(INPUT_Absolute_11 + INPUT_DIR)  # テキストファイルリスト
os.makedirs(OUTPUT_DIR_11 + INPUT_DIR,exist_ok=True)
nn=0
for text in textlist: #
            nn+=1
            if "DS_" in text:
                continue
            textname=INPUT_Absolute_11+INPUT_DIR+'/'+text
            li=[]
            numbers=[]

            with open(textname, mode = 'r') as t:
                numbers = [float(num) for num in t.read().split()] #テキストファイル内のポイントを切り取り
            row=0
            wk=len(numbers)//42
            #print(nn)
            #print(len(numbers))
            landmark_frame=[]
            for i in range(0,wk):
                landmark_frame=(numbers[row:row+42])
                row += 42
                #print(landmark_frame)
                li.extend(landmark_frame)
                li.extend(zero)
            mapped = ' '.join(map(str, li))
            with open(OUTPUT_DIR_11+INPUT_DIR+'/'+text, mode='w') as f:
                f.writelines(mapped)
```
